    Bucket age into groups (e.g. 0-18, 19-35, 36-50, 51-65, 65+). Which age group has the highest readmission rate?


```python
age_bins = [0, 18, 35, 50, 65, np.inf]
age_labels = ['0-18', '19-35', '36-50', '51-65', '65+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=True)

readmission_rates_by_age = df.groupby('age_group')['label'].mean().sort_values(ascending=False)

print("Readmission rates by age group:")
print(readmission_rates_by_age)

highest_readmission_group = readmission_rates_by_age.index[0]
highest_rate = readmission_rates_by_age.iloc[0]

print(f"\nThe age group with the highest readmission rate is {highest_readmission_group} with a rate of {highest_rate:.2%}.")
```

Which numeric columns are most correlated with the readmission target? :

```python
numeric_cols = df.select_dtypes(include=np.number).columns
correlation_matrix = df[numeric_cols].corr()['label'].sort_values(ascending=False)

print("Correlation of numeric columns with 'label' (Readmission Target):\n")
print(correlation_matrix)
```
