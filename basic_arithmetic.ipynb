```python
# Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.

# Note that the code stub provided reads from STDIN and passes arguments to the is_leap function. It is only necessary to complete the is_leap function.
```

```python
def is_leap(year):
    # A year is a leap year if it is divisible by 400
    if year % 400 == 0:
        return True
    # Else, if it is divisible by 100, it is NOT a leap year
    elif year % 100 == 0:
        return False
    # Else, if it is divisible by 4, it IS a leap year
    elif year % 4 == 0:
        return True
    # Otherwise, it is not a leap year
    else:
        return False



year = int(input())
print(is_leap(year))
```

```python
###n + 1 makes the loop include n.

# range(1, n + 1) generates numbers starting at 1 up to but not including n + 1.
# Since range stops before the second number, this means it goes through:

# 1, 2, 3, …, n

# So:

# If n = 5, range(1, 6) → 1, 2, 3, 4, 5
# If you used range(1, n), it would stop at n–1
# n + 1 is just how we tell Python to include n in the loop.
```

```python
n = int(input())

for i in range(1, n + 1):
    print(i, end='')
```

```python
n = int(input())

for i in range(1, n):
    print(i, end='')
```
