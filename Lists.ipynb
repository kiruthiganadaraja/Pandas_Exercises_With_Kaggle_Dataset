```python
my_list = [10, 20, 30, 40, 50]
print("Original list:", my_list)

# Pop without an index removes and returns the last item
popped_item_default = my_list.pop()
print("List after pop() (default, last item):", my_list)
print("Popped item (default):", popped_item_default)

# Pop with an index removes and returns the item at that index
popped_item_index = my_list.pop(1) # Removes item at index 1 (which is 20)
print("List after pop(1):", my_list)
print("Popped item at index 1:", popped_item_index)
```

### Difference Between `pop()` and `remove()`

-   **`pop(index)`**: Removes the item at the given index and returns it. If no index is provided, it removes and returns the last item in the list.
-   **`remove(value)`**: Removes the first occurrence of the specified value from the list. It does not return any value.

```python
my_list_pop = ['apple', 'banana', 'cherry', 'date']
print(f"Original list for pop(): {my_list_pop}")

# Using pop() without an index
popped_item = my_list_pop.pop() # Removes 'date'
print(f"List after pop() (last item): {my_list_pop}")
print(f"Item popped by pop(): {popped_item}")

# Using pop() with an index
popped_item_at_index = my_list_pop.pop(1) # Removes 'banana'
print(f"List after pop(1): {my_list_pop}")
print(f"Item popped by pop(1): {popped_item_at_index}")

print("\n---")

my_list_remove = ['red', 'green', 'blue', 'green', 'yellow']
print(f"Original list for remove(): {my_list_remove}")

# Using remove() with a value
my_list_remove.remove('green') # Removes the first 'green'
print(f"List after remove('green'): {my_list_remove}")

# If the value is not in the list, remove() raises a ValueError
try:
    my_list_remove.remove('orange')
except ValueError as e:
    print(f"Error when trying to remove 'orange': {e}")
```

```python
# Problem: Find the First Non-Repeating Element

# Given a list of integers, find the first element that appears only once.

# If every element repeats, print "NO UNIQUE ELEMENT".
```

```python
n = int(input()) # Read the number of elements
arr = list(map(int, input().split())) # Read the list of integers

found_unique = False
for element in arr:
    if arr.count(element) == 1:
        print(element)
        found_unique = True
        break # Stop after finding the first non-repeating element

if not found_unique:
    print("NO UNIQUE ELEMENT")
```

```python
# Python List

# A list can store different data types together.

# Example:

# arr = [1, "hello", 3.5, True]

# Lists are:

# Flexible
# Easy to use
# Most common in Python

# Created using:

# []
# Python Array

# Arrays are more restricted.

# They usually store only one data type.

# Example:

# from array import array

# arr = array('i', [1, 2, 3, 4])

# Here:

# 'i'

# means integer array.

# You cannot store strings or floats inside this integer array.

# Main Differences
# Feature	List	Array
# Data Types	Multiple types allowed	Usually one type only
# Flexibility	High	Lower
# Speed	Slightly slower	Faster for numeric data
# Memory Usage	More memory	Less memory
# Built-in	Yes	Need array module
# Mostly Used For	General programming	Numerical operations
# Example
# List
# x = [1, "hello", 3.5]

# Valid ✅

# Array
# from array import array

# x = array('i', [1, 2, 3])

# Valid ✅

# But this is invalid ❌:

# x = array('i', [1, "hello"])

# because arrays expect the same type.

# Important

# In coding interview problems and platforms like HackerRank or LeetCode:

# When they say:

# "array"

# they usually mean a Python list:

# arr = [1, 2, 3]

# not the actual array module.
```
