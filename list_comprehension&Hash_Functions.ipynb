````python
# Function to check if a number is a palindrome
##A palindrome number is a number that remains the same when its digits are reversed. For example, 121 is a palindrome number because if you reverse its digits, you still get 121. Other examples include 1, 11, 22, 101, 454, etc.
def is_palindrome(num):
    return str(num) == str(num)[::-1]

# Build a list of palindrome numbers within a range (e.g., up to 1000)
palindrome_numbers = [num for num in range(1, 1001) if is_palindrome(num)]
print(palindrome_numbers)
````

Hash Functions:

A **hash function** is a mathematical function that takes an input (data of any size) and produces a fixed-size output called a **hash value**, **hash code**, or **digest**.

For example:

Input:

```text
hello
```

Hash output:

```text
2cf24dba5fb0a...
```

The same input always produces the same hash.

---

## Simple idea

Think of a hash function like:

* a **digital fingerprint** for data.

Even a tiny change in input creates a very different hash.

Example:

```text
hello
→ abc123

hello!
→ x9k72p
```

---

## Important properties of a good hash function

### 1. Deterministic

Same input → same output every time.

### 2. Fast to compute

Should generate hashes quickly.

### 3. Fixed-size output

No matter how big the input is, output size stays constant.

Example:

* SHA-256 always gives 256-bit output.

### 4. Avalanche effect

Small input change → completely different hash.

### 5. Hard to reverse

You should not be able to reconstruct the original input from the hash.

---

## Common uses of hash functions

### Password storage

Instead of storing passwords directly:

```text
password123
```

systems store:

```text
ef92b778bafe...
```

When you log in, the system hashes your entered password and compares hashes.

---

### Data integrity

Used to check whether data changed.

Example:

* downloading a file,
* verifying software,
* blockchain verification.

If hashes differ, data was modified.

---

### Hash tables / dictionaries

Programming languages use hashes for fast lookup.

Example:

```python
my_dict["name"]
```

Internally uses hashing.

---

### Cryptography & Blockchain

Used in:

* Bitcoin
* digital signatures
* certificates
* secure communication

---

```text
     ↓
```


---

## Important distinction

### Hashing ≠ Encryption

| Hashing                             | Encryption           |
| ----------------------------------- | -------------------- |
| One-way                             | Two-way              |
| Cannot easily recover original data | Can decrypt with key |
| Used for verification               | Used for secrecy     |



One commonly asked interview concept is:

> Why do hash collisions happen?

Because infinite possible inputs are mapped into a finite output size, two different inputs can theoretically produce the same hash.

That’s called a **collision**. Good hash functions make collisions extremely rare.


Yes — absolutely. Python also uses hashing extensively, especially in:

* dictionaries (`dict`)
* sets (`set`)
* caching
* object lookup

And Python also has to handle hash collisions internally.

---

# Python dictionaries use hashing

Example:

```python
data = {
    "name": "John",
    "age": 30
}
```

When you do:

```python
data["name"]
```

Python:

1. computes the hash of `"name"`
2. maps it to a memory location (bucket)
3. retrieves the value quickly

That’s why dictionary lookup is usually:

```text
O(1)
```

average time complexity.

# Hash collisions happen in Python too

Two different objects can theoretically produce the same hash.

Python handles collisions internally using:

* open addressing,
* probing techniques,
* and equality checks.

---

# Important concept in Python

## Same hash does NOT mean same object

Python always verifies equality too.

Example:

```python
class Test:
    def __hash__(self):
        return 1

    def __eq__(self, other):
        return False

a = Test()
b = Test()

print(hash(a))
print(hash(b))
```

Both hashes:

```text
1
```

But:

```python
a != b
```

Python can still store both in a dictionary safely.

---

# Example collision handling

```python
class Person:
    def __init__(self, name):
        self.name = name

    def __hash__(self):
        return 42   # forced collision

    def __eq__(self, other):
        return self.name == other.name

p1 = Person("Alice")
p2 = Person("Bob")

d = {}
d[p1] = "Engineer"
d[p2] = "Manager"

print(d[p1])
print(d[p2])
```

Even though both hashes are `42`, Python still stores/retrieves correctly.

Why?
Because after hash match:

* Python checks `__eq__()`.

---

# Python hash table collision strategy

CPython mainly uses:

* open addressing,
* probing,
* sparse arrays.

It is highly optimized.

---

# Important interview point

## Mutable objects are usually NOT hashable

Example:

```python
hash([1,2,3])
```

Error:

```text
TypeError: unhashable type: 'list'
```

Because mutable objects can change after hashing, which would break dictionary behavior.

Immutable objects are hashable:

* strings
* tuples
* integers
* frozensets

---

# Another important Python detail

## Hash randomization

Python intentionally randomizes string hashes between runs for security reasons.

So:

```python
hash("hello")
```

may produce different values each time you restart Python.

This prevents certain denial-of-service collision attacks.

to make dictionaries and sets efficient.
